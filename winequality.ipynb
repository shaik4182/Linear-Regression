```python
y = winequality['quality']
```

```python
wine=winequality.drop(['quality'],axis=1)
```

```python
wine.head(2)
```

```python
x = wine
```

```python
x
```

```python
plt.figure(figsize=(15,10))
plt.tight_layout()
sns.distplot(winequality['quality'])
```

```python
regrssor = LinearRegression()
regrssor.fit(x,y)
```

```python
regrssor.coef_
```

```python
coff_ = pd.DataFrame(regrssor.coef_,x.columns,columns=['slope'])
```

```python
coff_
```

```python
intercept_=pd.DataFrame(regrssor.intercept_,x.columns,columns=['intercept'])
```

```python
intercept_
```

```python
y_pred = regrssor.predict(x)
```

```python
y_pred
```

```python
df = pd.DataFrame({'Actual':y,'Predicted':y_pred})
df
```

```python
df1 = df.head(25)
df1.plot(kind='bar',figsize=(10,8))
plt.grid(which='major',linestyle='-',linewidth='0.5',color='green')
plt.show()
```

```python
for i in x.columns:
    plt.scatter(x[i],y)
    plt.plot(x,regrssor.intercept_+regrssor.coef_*x)
    plt.show()
```

```python
for i in x.columns:
    plt.scatter(x[i],y)
    plt.plot(x,y_pred)
    plt.show()
```
